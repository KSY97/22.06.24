```python
# 3 + 4 -> 7 계산기 프로그램 작성
# 반복해서 작성
# 첫 문자에 'q'가 입력되면 프로그램 종료
```

```python
input_list = []

while True:
    input_data = input("계산식을 입력하세요(ex:10 + 20)(q를 누르면 종료) : ")
    input_list = input_data.split()
    
    if input_list[0] == 'q':
        print("프로그램을 종료합니다.")
        break
    
    x = int(input_list[0])
    y = int(input_list[2])

    if input_list[1] == '+':
        print("{} {} {} = {}".format(x, "+", y, x+y))
    elif input_list[1] == '-':
        print("{} {} {} = {}".format(x, "-", y, x-y))
    elif input_list[1] == '*':
        print("{} {} {} = {}".format(x, "*", y, x*y))
    elif input_list[1] == '/':
        print("{} {} {} = {}".format(x, "/", y, x/y))
    else:
        print("잘못 입력하셨습니다.")
```
